```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
resume_data = pd.read_csv('ResumeDataSet/UpdatedResumeDataSet.csv')
print(resume_data)
```

```python
X = resume_data["Category"].value_counts().index
y = resume_data["Category"].value_counts().values
```

```python
plt.figure(figsize=(12, 6))
plt.bar(X, y)
plt.title('Distribution of Resume Categories', fontsize=15)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Resumes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
